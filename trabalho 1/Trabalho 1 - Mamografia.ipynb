```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('.//mama.jpg')
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.gray()
```

```python
plt.imshow(img)
```

```python
kernel = np.ones((5,5),np.uint8)
```

```python
abertura = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
```

```python
# Abertura

plt.imshow(abertura)
```

```python
top_hat_abertura = cv2.subtract(img, abertura)
```

```python
# Top Hat Abertura

plt.imshow(top_hat_abertura)
```

```python
fechamento = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
```

```python
# Fechamento

plt.imshow(fechamento)
```

```python
top_hat_fechamento = cv2.subtract(fechamento, img)
```

```python
# Top Hat Fechamento

plt.imshow(top_hat_fechamento)
```

```python
contraste_morfologico = cv2.subtract(cv2.add(img, top_hat_abertura), top_hat_fechamento)
```

```python
# Contraste Morfolgico

plt.imshow(contraste_morfologico)
```

```python
remocao_fundo = cv2.subtract(contraste_morfologico, img)
```

```python
# Aps a remoo do fundo

plt.imshow(remocao_fundo)
```

```python
ret,threshold = cv2.threshold(remocao_fundo, 25, 255, cv2.THRESH_BINARY)
```

```python
# Resultado final.

plt.imshow(threshold)
```
